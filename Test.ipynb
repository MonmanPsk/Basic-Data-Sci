```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv("data-firearm.csv")
df.fillna(0, inplace=True)
state = []
year = []
gun = []

for i in df.index:
  if df["state"].duplicated()[i] == False:
    state.append(df["state"][i])
  else:
    continue

for i in df.index:
  if year.count(df["month"][i][:4]) == 0:
    year.append(df["month"][i][:4])

for i in state:
  gun.clear()
  for j in year:
    x = 0
    y = 0
    for k in df.index:
      if df["state"][k] == i and df["month"][k][:4] == j:
        x += df["handgun"][k]
        x += df["long_gun"][k]
        x += df["other"][k]
        if df["handgun"][k] > 0:
          y += 1
        if df["long_gun"][k] > 0:
          y += 1
        if df["other"][k] > 0:
          y += 1
    gun.append(x/y)
  plt.figure(figsize=(25,4))
  plt.plot(year, gun)
  plt.title(i)
  plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("steam.csv")
platform = {}
year = {}

for i in range(1):
  x = []
  y = []

  for j in df.index:
    if df["platforms"].duplicated()[j] == False:
      x.append(df["platforms"][j])
      y.append(1)
    else:
      y[x.index(df["platforms"][j])] += 1

  for j in range(len(x)):
    platform.update({x[j]:y[j]})

for i in range(1):
  x = []
  y = []

  for i in df.index:
    if x.count(df["release_date"][i][:4]) == 0:
      x.append(df["release_date"][i][:4])
  x.sort()

  for i in x:
    count = 0
    for j in df["release_date"]:
      if j[:4] == i:
        count += 1
    y.append(count)
  
  for i in range(len(x)):
    year.update({x[i]:y[i]})

plt.pie(list(platform.values()), labels=list(platform.keys()))
plt.title("Platforms Charts")
plt.show()

plt.figure(figsize=(25,5))
plt.bar(list(year.keys()), list(year.values()))
plt.title("Year Statistics")
plt.show()
```
